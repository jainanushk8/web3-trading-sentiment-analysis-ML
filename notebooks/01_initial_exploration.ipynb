```python
# Fixed Visualization Based on Actual Columns - UPDATED VERSION
import matplotlib.pyplot as plt
import seaborn as sns

# Your actual column names
actual_columns = list(trading_df.columns)
print(f"✅ Working with these {len(actual_columns)} columns: {actual_columns}")
print()

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Trade Direction using 'Side' column (BUY/SELL)
if 'Side' in trading_df.columns:
    side_counts = trading_df['Side'].value_counts()
    colors = ['#ff6b6b', '#4ecdc4']  # Red for SELL, Teal for BUY
    axes[0, 0].pie(side_counts.values, labels=side_counts.index, autopct='%1.1f%%', 
                   colors=colors, startangle=90)
    axes[0, 0].set_title('🔄 Trade Side Distribution (BUY vs SELL)')
else:
    axes[0, 0].text(0.5, 0.5, 'Side column not found', ha='center', va='center')
    axes[0, 0].set_title('Trade Direction Analysis')

# Plot 2: Top Trading Coins using 'Coin' column
if 'Coin' in trading_df.columns:
    top_coins = trading_df['Coin'].value_counts().head(10)
    bars = axes[0, 1].bar(range(len(top_coins)), top_coins.values, color='skyblue')
    axes[0, 1].set_title('🪙 Top 10 Trading Coins')
    axes[0, 1].set_xticks(range(len(top_coins)))
    axes[0, 1].set_xticklabels(top_coins.index, rotation=45)
    axes[0, 1].set_ylabel('Number of Trades')
    
    # Add value labels on bars
    for i, bar in enumerate(bars):
        height = bar.get_height()
        axes[0, 1].text(bar.get_x() + bar.get_width()/2., height + 0.01*max(top_coins.values),
                       f'{int(height)}', ha='center', va='bottom', fontsize=8)
else:
    axes[0, 1].text(0.5, 0.5, 'Coin column not found', ha='center', va='center')
    axes[0, 1].set_title('Trading Symbols Analysis')

# Plot 3: Closed PnL Distribution using 'Closed PnL' column
if 'Closed PnL' in trading_df.columns:
    pnl_data = pd.to_numeric(trading_df['Closed PnL'], errors='coerce').dropna()
    if len(pnl_data) > 0:
        # Create histogram with different colors for profit/loss
        axes[1, 0].hist(pnl_data, bins=50, alpha=0.7, color='lightgreen', edgecolor='black')
        axes[1, 0].axvline(x=0, color='red', linestyle='--', linewidth=2, label='Break-even')
        axes[1, 0].set_title('💰 Profit & Loss Distribution')
        axes[1, 0].set_xlabel('Closed PnL ($)')
        axes[1, 0].set_ylabel('Frequency')
        axes[1, 0].legend()
        
        # Add statistics text
        profit_trades = (pnl_data > 0).sum()
        total_trades = len(pnl_data)
        win_rate = (profit_trades / total_trades) * 100 if total_trades > 0 else 0
        axes[1, 0].text(0.02, 0.98, f'Win Rate: {win_rate:.1f}%\nTotal PnL: ${pnl_data.sum():.2f}', 
                       transform=axes[1, 0].transAxes, verticalalignment='top',
                       bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    else:
        axes[1, 0].text(0.5, 0.5, 'Closed PnL column exists but no valid numeric data', 
                       ha='center', va='center')
else:
    axes[1, 0].text(0.5, 0.5, 'Closed PnL column not found', ha='center', va='center')
    axes[1, 0].set_title('PnL Analysis')

# Plot 4: Top Active Traders using 'Account' column
if 'Account' in trading_df.columns:
    account_trades = trading_df['Account'].value_counts().head(15)
    bars = axes[1, 1].bar(range(len(account_trades)), account_trades.values, color='coral')
    axes[1, 1].set_title('👥 Top 15 Most Active Traders')
    axes[1, 1].set_xlabel('Trader Rank')
    axes[1, 1].set_ylabel('Number of Trades')
    
    # Shorten account addresses for display
    short_labels = [f'...{addr[-6:]}' for addr in account_trades.index]
    axes[1, 1].set_xticks(range(len(account_trades)))
    axes[1, 1].set_xticklabels(short_labels, rotation=45, fontsize=8)
    
    # Add grid for better readability
    axes[1, 1].grid(axis='y', alpha=0.3)
else:
    axes[1, 1].text(0.5, 0.5, 'Account column not found', ha='center', va='center')
    axes[1, 1].set_title('Trader Analysis')

plt.tight_layout()
plt.show()

# Enhanced column analysis
print("\n🔍 **DETAILED COLUMN ANALYSIS:**")
print("=" * 80)
for i, col in enumerate(actual_columns, 1):
    unique_vals = trading_df[col].nunique()
    data_type = trading_df[col].dtype
    sample_val = str(trading_df[col].iloc[0]) if len(trading_df) > 0 else "No data"
    
    # Add helpful insights for key columns
    insights = ""
    if col == 'Side':
        insights = " <- Trade direction (BUY/SELL)"
    elif col == 'Coin':
        insights = " <- Cryptocurrency symbols"
    elif col == 'Closed PnL':
        insights = " <- Profit/Loss in USD"
    elif col == 'Account':
        insights = " <- Trader wallet addresses"
    elif col == 'Size USD':
        insights = " <- Trade volume in dollars"
    elif 'Timestamp' in col:
        insights = " <- Time information"
    
    print(f"{i:2d}. {col:<20} | {str(data_type):<12} | {unique_vals:>7,} unique | {sample_val[:25]:<25}{insights}")
```
